```python
train_image_generator = tf.keras.preprocessing.image.ImageDataGenerator(rescale=1/255,rotation_range=45,
                                                                        zoom_range=0.2,
                                                                        horizontal_flip=True,
                                                                        width_shift_range=0.2,
                                                                        height_shift_range=0.2)
train_data_generator= train_image_generator.flow_from_directory(directory=train_dir,batch_size=100,
                                                                target_size=(150,150),
                                                                class_mode="binary")
```

```python
train_data_generator[0][0][0]
plt.imshow(train_data_generator[0][0][1])
```

```python
val_image_generator = tf.keras.preprocessing.image.ImageDataGenerator(rescale=1/255)
```

```python
val_data_generator= train_image_generator.flow_from_directory(directory=val_dir,batch_size=100,
                                                                target_size=(150,150),
                                                                class_mode="binary")
```

```python
#BUILDING CNN MODEL
```

```python
cnn1 = tf.keras.Sequential()
cnn1.add(tf.keras.layers.Conv2D(32,kernel_size=(3,3),activation='relu',input_shape=(150,150,3)))
cnn1.add(tf.keras.layers.MaxPooling2D())
cnn1.add(tf.keras.layers.Conv2D(64,kernel_size=(3,3),activation='relu'))
cnn1.add(tf.keras.layers.MaxPooling2D())
cnn1.add(tf.keras.layers.Conv2D(128,kernel_size=(3,3),activation='relu'))
cnn1.add(tf.keras.layers.MaxPooling2D())
cnn1.add(tf.keras.layers.Dropout(0.5))  
cnn1.add(tf.keras.layers.Flatten())
cnn1.add(tf.keras.layers.Dense(512,activation='relu'))
cnn1.add(tf.keras.layers.Dense(2,activation='softmax'))
```

```python
cnn1.summary()
```

```python
 cnn1.compile(loss=tf.keras.losses.SparseCategoricalCrossentropy(),
             optimizer='adam',
             metrics=['accuracy'])
```

```python
early_stop= tf.keras.callbacks.EarlyStopping(monitor='val_accuracy',patience=10)
batch_size=100       
model_fit = cnn1.fit(train_data_generator,validation_data=val_data_generator,epochs=100,
                      steps_per_epoch=total_train/batch_size,validation_steps=total_val/batch_size,    
                     callbacks=[early_stop])                                                        
                                                                                                                      
```

```python
import matplotlib.pyplot as plt
plt.plot(model_fit.history['loss'])
plt.plot(model_fit.history['val_loss'])
```
